## 課題1

各問で示されている図を出力してください。

その際、以下の要件を必ず満たすようにしてください。


1.   for文を必ず用いること
2.   変数`size`の値を変えることで自由に図形の大きさを変更できること




### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
size = 5
for i in range(1, size + 1):
    print("*" * i)
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
size = 5
for i in range(size):
    print(" " * i + "*")
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 5
for i in range(size):
    print(" " * (size - i - 1) + "*")
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 5
for i in range(size):
    row = ""
    for j in range(size):
        if j == i or j == size - i - 1:
            row += "*"
        else:
            row += " "
    print(row)
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
size = 5
for i in range(size):
    spaces = " " * (size - i - 1)
    stars = "*" * (2 * i + 1)
    print(spaces + stars)
````
